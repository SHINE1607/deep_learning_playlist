```
from keras.layers import Input, Dense, Flatten 
from keras.models import Model
#these are classification models in keras 
#however we need to create our last layer to our model
from keras.applications.vgg16 import VGG16 
from keras.applications.vgg16 import preprocess_input
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator 
from keras.models import Sequential
#in VGG16 has an putpur layer has of 1000 diferent layer output
#so we will drop the last layer 
import numpy as np
from glob import glob 
import matplotlib.pyplot as plt
```

```
#vgg16 was modelled such a way that input image was size of 224, 224 
image_size = [224, 224]

train_path  = "datasets/Train"
valid_path = "datasets/Test"

```

```
#add preprocessing layer to vgg16 
#input_shape in vgg16 has to be [224, 224, 3]
vgg = VGG16(input_shape = image_size + [3], weights = "imagenet", include_top = False)
#inlcude_top = False ==> drops the last layer from the model
```

```
#we dont have to train vgg16 layers 
for layer in vgg.layers:
  layer.trainable = False
```

```
folders = glob("datasets/train/")
```

```
#x is retriecving the output layer from the model and instantiate a new model
x = Flatten()(vgg.output)
#creating the last layer if vgg model
prediction_layer = Dense(len(folders), activation = "softmax")(x)
model = Model(inputs = vgg.inputs, outputs = prediction_layer)
```

```

model.summary()
```

```
model.compile(loss = "categorical_crossentropy", 
              optimizer = "adam", 
              metrica = ["accuracy"])
```
